```python
from notebook_resolver import *
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("results/method_comparison.xlsx", sheet_name="Sheet1")

# Convert numeric columns
df["R2"] = pd.to_numeric(df["R2"], errors="coerce").clip(lower=0)  # R² should be non-negative
df["MSE"] = pd.to_numeric(df["MSE"], errors="coerce").clip(lower=0)  # MSE should be non-negative
df["MAE"] = pd.to_numeric(df["MAE"], errors="coerce").clip(lower=0)  # MAE should be non-negative

# Methods order
methods = ["IT2TSK", "MLP", "SVR", "RF", "KNN"]

# Unique datasets
datasets = df["Worksheet Name"].unique()

# Line Chart for R²
plt.figure(figsize=(14,6))
for method in methods:
    subset = df[df["Predictor Method"] == method]
    plt.plot(subset["Worksheet Name"], subset["R2"], marker="o", label=method)

plt.title("Comparison of R² Across Methods", fontsize=18, fontweight='bold')
plt.ylabel("R²", fontsize=14)
plt.xticks(rotation=75)
plt.legend()
plt.tight_layout()
plt.show()

# Line Chart for MSE
plt.figure(figsize=(14,6))
for method in methods:
    subset = df[df["Predictor Method"] == method]
    plt.plot(subset["Worksheet Name"], subset["MSE"], marker="o", label=method)

plt.title("Comparison of MSE Across Methods", fontsize=18, fontweight='bold')
plt.ylabel("MSE", fontsize=14)
plt.xticks(rotation=75)
plt.yscale("log")  # log scale for readability
plt.legend()
plt.tight_layout()
plt.show()

# Line Chart for MAE
plt.figure(figsize=(14,6))
for method in methods:
    subset = df[df["Predictor Method"] == method]
    plt.plot(subset["Worksheet Name"], subset["MAE"], marker="o", label=method)

plt.title("Comparison of MAE Across Methods", fontsize=18, fontweight='bold')
plt.ylabel("MAE", fontsize=14)
plt.xticks(rotation=75)
plt.yscale("log")  # log scale for readability
plt.legend()
plt.tight_layout()
plt.show()
```
