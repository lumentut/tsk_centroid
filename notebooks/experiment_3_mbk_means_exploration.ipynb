### Experiment #3 (MBK-Means Exploration)
- Script: it2tsk_mbk_exploration.py
- Raw data generation command:
```bash
	$ bin/it2tsk_mbk_exploration
```
- Raw data result: IT2TSK_MBKMeans_Exploration.xlsx
- Time required: total +/- 12 minutes (12 sheets x 27 configuration (3 × 3 × 3))

````python
from notebook_resolver import *
import pandas as pd
from src.utils.excel_operation import export_to_excel

file_path = "experiments/IT2TSK_MBKMeans_Exploration.xlsx"

df = pd.read_excel(file_path)
df["Sheet_Order"] = df["Sheet Name"].str.extract(r"^(\d+)").astype(int)
df_max = df.loc[df.groupby("Sheet Name")["R2"].idxmax()]
df_max[[
    "R2", 
    "Rule Base Clustering Time", 
    "Features Clustering Time",
    "Clustering Time"]] = df_max[[
    "R2", 
    "Rule Base Clustering Time", 
    "Features Clustering Time",
    "Clustering Time"]].round(4)
df_sorted = df_max.sort_values("Sheet_Order").reset_index(drop=True)
selected_max_df = df_sorted[[
    "Sheet Name", 
    "R2",
    "Rule Base Clustering Time",
    "Features Clustering Time",
	"Clustering Time",
	"batch_size",
    "tol",
    "max_no_improvement"]]

# Compute average for selected numeric columns
avg_row = pd.DataFrame({
    "Sheet Name": ["Average"],
    "R2": [df_sorted["R2"].mean()],
    "Rule Base Clustering Time": [df_sorted["Rule Base Clustering Time"].mean()],
    "Features Clustering Time": [df_sorted["Features Clustering Time"].mean()],
    "Clustering Time": [df_sorted["Clustering Time"].mean()],
    "batch_size": [""],
    "tol": [""],
    "max_no_improvement": [""]
}).round(4)

final_df = pd.concat([selected_max_df, avg_row], ignore_index=True)

export_to_excel(final_df, file_path, "best_params")

final_df
````
