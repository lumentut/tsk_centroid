```python
import pandas as pd
from notebook_resolver import *
from src.utils.pandas_extension import *
from src.dataset import Dataset, WorkSheet

dataset_path = "data/e-nose_dataset_12_beef_cuts.xlsx"

data = {}

for worksheet in WorkSheet:
    dataset = Dataset(path=dataset_path, sheet_name=worksheet.value)
    statistics_df = dataset.train_df.stats.describe()
    data[worksheet.value] = statistics_df.to_dict()

for ws in data:
	print(f"Statistics for {ws}:")
	stats_df = pd.DataFrame(data[ws])
	display(stats_df)
	print("\n" + "="*80 + "\n")
```

```python
data_stats = []
for ws in data:
	# display(data[ws])
	for k1, v1 in data[ws].items():
		stats = {}
		for k, v in v1.items():
			stats['Sheet Name'] = ws
			stats['Feature'] = k1
			stats[k] = v
		data_stats.append(stats)
  
	
stats_df = pd.DataFrame(data_stats)
stats_df
```
