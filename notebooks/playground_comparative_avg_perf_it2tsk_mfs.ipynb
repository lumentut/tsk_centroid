### Comparative Average Perfomance IT2TSK Triangular, Trapezoidal, and Gaussian

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the three metrics
membership_functions = ['Trapezoidal', 'Triangular', 'Gaussian']

# Performance data (approximate values from the chart)
r2_values = [0.71, -0.04, 0.98]
mse_values = [0.024, 0.090, 0.002]
mae_values = [0.092, 0.142, 0.024]

# Create figure with three subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Comparative Performance of Default IT2TSK Models Using Different MFs', 
             fontsize=14, fontweight='bold', y=1.02)

# Different colors for each membership function
colors = ['#FF6B35', '#4ECDC4', '#45B7D1']  # Orange, Teal, Blue
# Alternative color schemes:
# colors = ['#E74C3C', '#2ECC71', '#3498DB']  # Red, Green, Blue
# colors = ['#F39C12', '#9B59B6', '#1ABC9C']  # Orange, Purple, Turquoise

# Plot R² values
bars1 = axes[0].bar(membership_functions, r2_values, color=colors, alpha=0.8)
axes[0].set_title('Average R²', fontsize=12, fontweight='bold')
axes[0].set_ylabel('R²', fontsize=11)
axes[0].set_ylim(-0.2, 1.05)
axes[0].grid(True, alpha=0.3)
axes[0].tick_params(axis='x', rotation=0)

# Plot MSE values
bars2 = axes[1].bar(membership_functions, mse_values, color=colors, alpha=0.8)
axes[1].set_title('Average MSE', fontsize=12, fontweight='bold')
axes[1].set_ylabel('MSE', fontsize=11)
axes[1].set_ylim(0, 0.09)
axes[1].grid(True, alpha=0.3)
axes[1].tick_params(axis='x', rotation=0)

# Plot MAE values
bars3 = axes[2].bar(membership_functions, mae_values, color=colors, alpha=0.8)
axes[2].set_title('Average MAE', fontsize=12, fontweight='bold')
axes[2].set_ylabel('MAE', fontsize=11)
axes[2].set_ylim(0, 0.15)
axes[2].grid(True, alpha=0.3)
axes[2].tick_params(axis='x', rotation=0)

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()

# Optional: Save the figure
# plt.savefig('it2tsk_performance_comparison.png', dpi=300, bbox_inches='tight')


```
